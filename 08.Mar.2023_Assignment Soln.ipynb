Q1. What is the Probability density function?

A **Probability Density Function (PDF)** is a function that describes the likelihood of a continuous random variable taking on a particular value. It is a fundamental concept in probability and statistics used to model continuous data. 

Here are the key points about a PDF:

### 1. **Definition**:
The PDF, denoted as \( f(x) \), is such that the probability of the random variable \( X \) falling within a specific range \([a, b]\) is given by the integral of the PDF over that range:

\[
P(a \leq X \leq b) = \int_a^b f(x) \, dx
\]

### 2. **Properties**:
- \( f(x) \geq 0 \): The PDF is always non-negative because probabilities cannot be negative.
- The total area under the PDF curve is 1, representing the certainty that the random variable takes some value in its range:
\[
\int_{-\infty}^\infty f(x) \, dx = 1
\]

### 3. **Interpretation**:
- The value of \( f(x) \) at a particular point does not give the probability of \( X = x \) (since for continuous variables, the probability of taking any exact value is 0).
- Instead, \( f(x) \) represents the relative likelihood or density of \( X \) near \( x \). 

### 4. **Examples**:
- **Uniform Distribution**: 
\[
f(x) =
\begin{cases} 
\frac{1}{b-a}, & a \leq x \leq b \\
0, & \text{otherwise}
\end{cases}
\]
This indicates equal probability density across the range \([a, b]\).

- **Normal Distribution (Gaussian)**:
\[
f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
\]
where \( \mu \) is the mean, and \( \sigma^2 \) is the variance. The PDF is bell-shaped.

### 5. **Applications**:
PDFs are widely used in statistics, machine learning, and data science to:
- Model distributions of continuous data.
- Compute probabilities over intervals.
- Formulate likelihood functions for parameter estimation.

Q2. What are the types of Probability distribution?

Probability distributions can be broadly categorized into **discrete** and **continuous** types based on the nature of the random variable they describe. Below are the main types of probability distributions:

---

### **1. Discrete Probability Distributions**
Discrete distributions apply to random variables that take on countable values (e.g., integers).

#### a. **Bernoulli Distribution**:
- Describes a random experiment with exactly two outcomes: success (1) and failure (0).
- Example: Tossing a coin (success = heads, failure = tails).
- Probability Mass Function (PMF):
\[
P(X = x) =
\begin{cases} 
p, & x = 1 \\
1 - p, & x = 0 
\end{cases}
\]
where \( p \) is the probability of success.

---

#### b. **Binomial Distribution**:
- Models the number of successes in \( n \) independent Bernoulli trials.
- Example: Flipping a coin 10 times and counting the number of heads.
- PMF:
\[
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}, \, k = 0, 1, \dots, n
\]
where \( n \) is the number of trials and \( p \) is the probability of success.

---

#### c. **Poisson Distribution**:
- Describes the number of events occurring in a fixed interval of time or space, assuming events occur independently.
- Example: Number of customer arrivals at a store in an hour.
- PMF:
\[
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \, k = 0, 1, 2, \dots
\]
where \( \lambda \) is the average rate of occurrence.

---

#### d. **Geometric Distribution**:
- Models the number of Bernoulli trials needed to get the first success.
- Example: Rolling a die until the first "6" appears.
- PMF:
\[
P(X = k) = (1-p)^{k-1} p, \, k = 1, 2, \dots
\]
where \( p \) is the probability of success.

---

### **2. Continuous Probability Distributions**
Continuous distributions apply to random variables that take on an infinite number of possible values within a range.

#### a. **Normal (Gaussian) Distribution**:
- Models many natural phenomena; its PDF is bell-shaped.
- Example: Heights, test scores, measurement errors.
- PDF:
\[
f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
\]
where \( \mu \) is the mean, and \( \sigma^2 \) is the variance.

---

#### b. **Uniform Distribution**:
- Describes a random variable equally likely to take any value in a specified range.
- Example: Rolling a fair die.
- PDF:
\[
f(x) =
\begin{cases} 
\frac{1}{b-a}, & a \leq x \leq b \\
0, & \text{otherwise}
\end{cases}
\]

---

#### c. **Exponential Distribution**:
- Models the time between events in a Poisson process.
- Example: Time until the next earthquake.
- PDF:
\[
f(x) = \lambda e^{-\lambda x}, \, x \geq 0
\]
where \( \lambda \) is the rate parameter.

---

#### d. **Gamma Distribution**:
- Generalizes the exponential distribution to model waiting times for multiple events.
- Example: Time until 3 phone calls are received.

---

### **3. Other Types of Distributions**
#### a. **Log-Normal Distribution**:
- A variable whose logarithm follows a normal distribution.
- Example: Stock prices.

#### b. **Beta Distribution**:
- Used for random variables in the range [0, 1]; often for modeling probabilities.

#### c. **Chi-Square Distribution**:
- Used in hypothesis testing and confidence intervals for variance.
- Example: Goodness-of-fit test.

---

### Choosing the Right Distribution
The choice depends on:
- Nature of the data (discrete vs. continuous).
- Characteristics like mean, variance, skewness.
- Underlying processes generating the data.

Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

````python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution.

    Parameters:
        x (float): The point at which to calculate the PDF.
        mean (float): The mean of the normal distribution (μ).
        std_dev (float): The standard deviation of the normal distribution (σ).

    Returns:
        float: The value of the PDF at x.
    """
    exponent = math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    denominator = std_dev * math.sqrt(2 * math.pi)
    return exponent / denominator

# Example usage:
mean = 0    # Mean (μ)
std_dev = 1 # Standard deviation (σ)
x = 1       # Point at which to calculate the PDF
pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF value at x={x} for N({mean}, {std_dev}^2) is {pdf_value}")
````

### Explanation of the Code:
- **Formula**:
  The PDF of a normal distribution is given by:
  \[
  f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
  \]
- **Parameters**:
  - \( x \): The point at which the PDF is evaluated.
  - \( \mu \) (`mean`): The mean of the distribution.
  - \( \sigma \) (`std_dev`): The standard deviation of the distribution.
- **Steps**:
  - Compute the exponent term: \( e^{-\frac{(x-\mu)^2}{2\sigma^2}} \).
  - Compute the denominator: \( \sigma \sqrt{2\pi} \).
  - Return the value of \( f(x) \) by dividing the exponent by the denominator.

### Example Output:
For \( N(0, 1^2) \) and \( x = 1 \):
\[
f(1) = \frac{1}{1 \cdot \sqrt{2\pi}} e^{-\frac{(1-0)^2}{2 \cdot 1^2}} = 0.24197
\]

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

### **Properties of Binomial Distribution**:

1. **Fixed Number of Trials**:
   - The experiment consists of \( n \) independent trials, where \( n \) is a fixed positive integer.

2. **Binary Outcomes**:
   - Each trial has only two possible outcomes: **success** (with probability \( p \)) and **failure** (with probability \( 1-p \)).

3. **Independent Trials**:
   - The outcome of one trial does not affect the outcome of any other trial.

4. **Constant Probability of Success**:
   - The probability of success (\( p \)) is the same for each trial.

5. **Random Variable**:
   - The random variable \( X \) represents the number of successes in \( n \) trials, and it can take values \( 0, 1, 2, \dots, n \).

6. **Probability Mass Function (PMF)**:
   - The probability of exactly \( k \) successes is given by:
   \[
   P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}, \, k = 0, 1, 2, \dots, n
   \]
   where \( \binom{n}{k} = \frac{n!}{k! (n-k)!} \) is the binomial coefficient.

7. **Mean, Variance, and Standard Deviation**:
   - Mean: \( \mu = n \cdot p \)
   - Variance: \( \sigma^2 = n \cdot p \cdot (1-p) \)
   - Standard Deviation: \( \sigma = \sqrt{n \cdot p \cdot (1-p)} \)

---

### **Examples of Events Where Binomial Distribution Can Be Applied**:

1. **Quality Control in Manufacturing**:
   - Suppose a factory produces light bulbs, and 95% of the bulbs are defect-free (\( p = 0.95 \)).
   - If 10 bulbs are randomly selected, the binomial distribution can be used to model the number of defect-free bulbs among the 10 trials.

2. **Survey Analysis**:
   - In a political survey, if 60% of people support a particular candidate (\( p = 0.6 \)), and 15 people are randomly surveyed, the binomial distribution can model the number of people who support the candidate.

---

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

Here’s a Python function to calculate the **Cumulative Distribution Function (CDF)** of a Poisson distribution at a given point \( k \), for a given mean \( \lambda \):

````python
import math

def poisson_cdf(mean, k):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution.

    Parameters:
        mean (float): The mean (λ) of the Poisson distribution.
        k (int): The point up to which the CDF is calculated (inclusive).

    Returns:
        float: The CDF value at k.
    """
    cdf = 0.0
    for i in range(0, k + 1):
        cdf += (math.exp(-mean) * (mean ** i) / math.factorial(i))
    return cdf

# Example usage:
mean = 3  # Mean (λ)
k = 5     # Point at which to calculate the CDF
result = poisson_cdf(mean, k)
print(f"The CDF of a Poisson distribution with mean {mean} at k={k} is {result}")
````

````python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf_optimized(mean, k):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution.

    Parameters:
        mean (float): The mean (λ) of the Poisson distribution.
        k (int): The point up to which the CDF is calculated (inclusive).

    Returns:
        float: The CDF value at k.
    """
    # Using scipy's poisson.cdf for optimized computation
    return poisson.cdf(k, mean)

def plot_poisson_cdf(mean, max_k):
    """
    Plot the Cumulative Distribution Function (CDF) of a Poisson distribution.

    Parameters:
        mean (float): The mean (λ) of the Poisson distribution.
        max_k (int): The maximum value of k to plot.
    """
    x_values = np.arange(0, max_k + 1)
    cdf_values = [poisson_cdf_optimized(mean, k) for k in x_values]

    plt.figure(figsize=(8, 6))
    plt.step(x_values, cdf_values, where="post", color="blue", label=f"Poisson CDF (λ={mean})")
    plt.title("Poisson Distribution CDF", fontsize=14)
    plt.xlabel("k (Number of occurrences)", fontsize=12)
    plt.ylabel("Cumulative Probability (F(k))", fontsize=12)
    plt.grid(alpha=0.4, linestyle="--")
    plt.legend(fontsize=12)
    plt.xticks(x_values)
    plt.show()

# Example usage:
mean = 3  # Mean (λ)
k = 5     # Point to calculate CDF
result = poisson_cdf_optimized(mean, k)
print(f"The CDF of a Poisson distribution with mean {mean} at k={k} is {result:.4f}")

# Plotting the CDF for visualization
plot_poisson_cdf(mean, 15)
````

### Explanation:
1. **Poisson Distribution**:
   The probability of exactly \( k \) occurrences is given by:
   \[
   P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
   \]
   where \( \lambda \) is the mean.

2. **Cumulative Distribution Function**:
   The CDF is the sum of probabilities up to and including \( k \):
   \[
   F(k) = P(X \leq k) = \sum_{i=0}^k \frac{\lambda^i e^{-\lambda}}{i!}
   \]

3. **Implementation Details**:
   - The loop iterates from \( i = 0 \) to \( i = k \) and accumulates the probabilities.
   - Uses Python’s `math.exp` for the exponential function and `math.factorial` for factorials.

Q7. How Binomial distribution different from Poisson distribution?

The **Binomial distribution** and the **Poisson distribution** are both discrete probability distributions, but they differ in their assumptions, use cases, and parameters. Here's a detailed comparison:

---

### **1. Nature of the Experiment**

- **Binomial Distribution**:
  - Describes the probability of a fixed number of **successes** in a fixed number of **independent trials**.
  - Each trial has two possible outcomes: **success** or **failure**.
  - Example: Flipping a coin \( n \) times and counting the number of heads.

- **Poisson Distribution**:
  - Describes the probability of a certain number of **events** occurring in a fixed interval of time or space.
  - Events occur randomly and independently.
  - Example: Number of customer arrivals at a store in an hour.

---

### **2. Parameters**

- **Binomial Distribution**:
  - Defined by two parameters:
    - \( n \): Number of trials.
    - \( p \): Probability of success in a single trial.

- **Poisson Distribution**:
  - Defined by one parameter:
    - \( \lambda \): The average rate of occurrence (mean number of events per interval).

---

### **3. Key Assumptions**

- **Binomial Distribution**:
  - Fixed number of trials (\( n \)).
  - Probability of success (\( p \)) is constant across trials.
  - Trials are independent.

- **Poisson Distribution**:
  - Events are independent.
  - The average rate of occurrence (\( \lambda \)) is constant over the interval.
  - Two events cannot occur at the same exact instant.

---

### **4. Support (Values of the Random Variable)**

- **Binomial Distribution**:
  - The number of successes (\( X \)) can take integer values from \( 0 \) to \( n \).

- **Poisson Distribution**:
  - The number of events (\( X \)) can take non-negative integer values from \( 0, 1, 2, \dots, \infty \).

---

### **5. Formula**

- **Binomial PMF**:
  \[
  P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}, \, k = 0, 1, \dots, n
  \]

- **Poisson PMF**:
  \[
  P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \, k = 0, 1, 2, \dots
  \]

---

### **6. Relationship Between the Distributions**

- The Poisson distribution can be derived as a **limiting case of the Binomial distribution**:
  - When \( n \to \infty \), \( p \to 0 \), and \( n \cdot p = \lambda \), the Binomial distribution approximates a Poisson distribution.
  - This happens in scenarios where the number of trials is very large, but the probability of success in each trial is very small (rare events).

---

### **7. Examples**

#### **Binomial Distribution**:
- Tossing a coin 10 times and counting the number of heads.
- Testing 20 light bulbs to count how many are defective.

#### **Poisson Distribution**:
- Counting the number of customers arriving at a store in an hour.
- Counting the number of emails received in a day.

---

### **Summary of Differences**

| Feature                | Binomial Distribution                     | Poisson Distribution              |
|------------------------|-------------------------------------------|-----------------------------------|
| Random Variable        | Number of successes                      | Number of events                 |
| Parameters             | \( n \) (trials), \( p \) (success prob.)| \( \lambda \) (average rate)     |
| Support                | \( 0 \leq X \leq n \)                    | \( 0, 1, 2, \dots, \infty \)     |
| Fixed Number of Trials | Yes                                       | No                                |
| Use Case               | Success/failure experiments              | Random events over time/space    |

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

The **mean** and **variance** in both the **Binomial distribution** and **Poisson distribution** have specific relationships, reflecting their respective properties.

### **1. Binomial Distribution**

For a **Binomial distribution**, which models the number of successes in \( n \) independent trials with success probability \( p \), the mean and variance are related as follows:

- **Mean** (\( \mu \)) of a Binomial distribution:
  \[
  \mu = n \cdot p
  \]
  - It represents the expected number of successes in \( n \) trials.

- **Variance** (\( \sigma^2 \)) of a Binomial distribution:
  \[
  \sigma^2 = n \cdot p \cdot (1 - p)
  \]
  - It reflects the spread of the number of successes across trials, with the variance being larger when \( p \) is closer to 0.5.

#### **Relationship**:
- The **mean** is the product of the number of trials \( n \) and the probability of success \( p \).
- The **variance** is also dependent on \( n \) and \( p \), but it is moderated by the factor \( (1 - p) \), reflecting the trade-off between success and failure probabilities.

### **2. Poisson Distribution**

For a **Poisson distribution**, which models the number of events occurring in a fixed interval of time or space with a constant mean rate of occurrence \( \lambda \), the mean and variance are related as follows:

- **Mean** (\( \mu \)) of a Poisson distribution:
  \[
  \mu = \lambda
  \]
  - It represents the expected number of events occurring in a given interval.

- **Variance** (\( \sigma^2 \)) of a Poisson distribution:
  \[
  \sigma^2 = \lambda
  \]
  - The variance in a Poisson distribution is equal to its mean, reflecting the fact that the distribution is defined by a single parameter \( \lambda \), which governs both the average and the spread of the data.

#### **Relationship**:
- In the Poisson distribution, **mean** and **variance** are **equal** and both equal to the rate parameter \( \lambda \). This indicates that as the average rate of occurrence increases, both the expected number of events and the variability of the number of events increase proportionally.

---

### **Summary of Relationships**

| Distribution       | Mean (\( \mu \))                     | Variance (\( \sigma^2 \))               |
|--------------------|--------------------------------------|----------------------------------------|
| **Binomial**       | \( \mu = n \cdot p \)                | \( \sigma^2 = n \cdot p \cdot (1 - p) \) |
| **Poisson**        | \( \mu = \lambda \)                  | \( \sigma^2 = \lambda \)               |

- **Binomial**: The variance depends on both \( n \) and \( p \), and is generally smaller than the mean when \( p \) is far from 0.5.
- **Poisson**: The mean and variance are equal and both determined by the rate \( \lambda \), indicating a direct relationship between the expected value and its spread.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a **Normal distribution**, the least frequent data points (i.e., the data points with the lowest probabilities) appear at the **tails** of the distribution, farthest from the **mean**.

### Key Points:
- A **Normal distribution** is symmetric around the **mean** (\( \mu \)).
- The **mean**, **median**, and **mode** are all located at the center of the distribution.
- As you move **away** from the mean, the frequency of data points **decreases**, approaching zero at the extreme ends (the **tails**).
  
### Least Frequent Data Points:
- The **least frequent** data points appear far from the **mean**, particularly in the **left** and **right tails**.
- The probability density function (PDF) of a Normal distribution decreases exponentially as you move away from the mean.

### Visualization:

- **Mean (µ)**: The peak of the bell curve, where the most frequent values occur.
- **Tails**: As you move further from the mean (either to the left or right), the frequency of values decreases.
  - **Far left or far right**: These regions contain the least frequent data points.

In a normal distribution, the data points **far away from the mean** (typically beyond 3 standard deviations) are the **least frequent**, as they lie in the areas where the probability density function approaches zero.

---

### Example:
- In a Normal distribution with \( \mu = 0 \) and \( \sigma = 1 \), the data points at \( x = 3 \) or \( x = -3 \) (i.e., 3 standard deviations away from the mean) are already very infrequent, and beyond these points, the frequency continues to decrease as we approach the tails.
