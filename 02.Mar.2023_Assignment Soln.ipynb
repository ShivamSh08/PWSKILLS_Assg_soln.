#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib?<br>
**Ans:**<br>
Matplotlib is a comprehensive library for creating static, interactive, and animated visualizations in Python. It is extensively used for creating 2D plots, which can be used for data analysis, data visualization, and scientific visualization, among other applications. Matplotlib is highly customizable and supports many different plot types and data formats. Matplotlib makes easy things easy and hard things possible.

* Create publication quality plots.
* Make interactive figures that can zoom, pan, update.
* Customize visual style and layout.
* Export to many file formats.

The Pyplot module is a part of Matplotlib that provides a MATLAB-like interface for creating plots. It is often used for creating simple plots quickly and easily. Here are five common plots that can be created using the Pyplot module:

1. Line plot: This is a basic plot that connects data points with a straight line. It is useful for showing trends or changes over time.

2. Scatter plot: This is a plot of points that are not connected by lines. It is useful for visualizing the relationship between two variables.

3. Histogram: This is a plot that shows the distribution of a single variable. It is useful for visualizing the frequency of different values in a dataset.

4. Bar plot: This is a plot that shows the relationship between two variables, where one variable is categorical and the other is numerical. It is useful for comparing different categories.

5. Pie chart: This is a plot that shows the proportions of different categories in a dataset. It is useful for visualizing the composition of a dataset.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.<br>

**Ans:**<br>
Scatter plots are used to observe relationships between variables and use dots to represent the relationship between them. The scatter() method in the matplotlib library is used to draw a scatter plot. Scatter plots are widely used to represent relations among variables and how change in one affects the other.<br>
Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a marker whose position depends on its values in the columns set on the X and Y axes.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(' X vs Y')
plt.scatter(x,y)
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.<br>Use the following data:<br>

import numpy as np<br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])<br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])<br>
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])<br>
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])<br>

**Ans:**<br>

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to arrange plots in a grid-like structure, with rows and columns, and you can specify the position of each plot using row and column indices. This is useful when you want to compare multiple plots side by side or when you want to create a complex plot with multiple subplots.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each line plot
data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
]

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2)

# Iterate over the data and create a line plot for each subplot
for i, (x, y) in enumerate(data):
    row = i // 2
    col = i % 2
    axs[row, col].plot(x, y)
    axs[row, col].set_title(f'Line {i+1}')

# Adjust the layout to prevent overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.<br>
import numpy as np<br>
company = np.array(["Apple", "Microsoft", "Google", "AMD"])<br>
profit = np.array([3000, 8000, 1000, 10000])<br>
**Ans:**<br>
Bar graphs are the pictorial representation of data, in the form of vertical or horizontal rectangular bars, where the length of bars are proportional to the measure of data. A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit in Billions')
plt.title('Company vs Profit')
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.<br>
box1 = np.random.normal(100, 10, 200)<br>
box2 = np.random.normal(90, 20, 200)<br>
**Ans:**<br>
Box Plot also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.Box Plots use quartiles to show how a given data is distributed.<br>
Boxplots pack more information in a very small face and can be useful when comparing two or more variables on a single plot.<br>
Usage of BOX plot is as follows:-
* It is used to represent the quartile/percentile distribution of data.
* It is used to visualize distribution of numerical data.
* It represents a 5 point summary, they are (Q1,Q2,Q3,min and max).

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.show()
```
